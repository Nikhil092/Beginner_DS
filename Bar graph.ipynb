```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data',index_col =0)
df
```

```python
df.sort_values(by = 'Very interested',ascending = False,inplace = True)
df = round(df*100/2233, 2)
```

```python
df
```

```python
color = ['#5cb85c','#5bc0de','#d9534f']
ax = df.plot(kind = 'bar',figsize = (20,8), color =color,width = .8, fontsize = 14)
ax.legend(loc = 'upper right',fontsize = 14)
ax.set_title('Percentages of Respondets\' interest in Data Science Topics',fontsize = 16)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(axis = 'y',left = False,  labelleft = False )

for i, v in enumerate(df['Very interested']):
    plt.text(i - .4, v + .5, str(v), color='black', fontsize=14)
    
for i, v in enumerate(df['Somewhat interested']):
    plt.text(i-.1, v + .5, str(v), color='black', fontsize=14)
    
for i, v in enumerate(df['Not interested']):
    plt.text(i + .175, v + .5, str(v), color='black', fontsize=14)    
```
